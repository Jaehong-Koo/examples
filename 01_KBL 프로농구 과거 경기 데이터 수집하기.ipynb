```python
## 동작 관련 함수
# 경기목록을 클릭하는 함수
def click_match_box():
    driver.find_element(By.XPATH, '//*[@id="container"]/main/div[3]/div/div[2]/div/div/select').click()

# 경기목록 안의 경기를 선택하는 함수
def click_match(game_num):
    driver.find_element(By.XPATH, '//*[@id="container"]/main/div[3]/div/div[2]/div/div/select/option['+str(game_num)+']').click()
    
# 매치박스안의 경기 수를 파악하는 함수
def match_len():
    match_length = len(driver.find_element(By.XPATH, '//*[@id="container"]/main/div[3]/div/div[2]/div/div/select').text.split('\n'))
    
    return match_length

# 다음 날짜 버튼을 클릭하는 함수
def click_next_day():
    driver.find_element(By.XPATH, '//*[@id="container"]/main/div[3]/div/div[1]/span/a[2]/i').click()


## 데이터 추출 관련 함수
# 날짜 추출하는 함수
def data_date():
    date = driver.find_element(By.XPATH, '//*[@id="container"]/main/div[3]/div/div[1]/span/span').text
    date_edited = date[:-4]

    return date_edited

# 팀 이름을 가져오는 함수(1:홈팀, 2:어웨이팀)
def data_team_name(team_num):
    team_name = driver.find_element(By.XPATH, '//*[@id="container"]/main/div[3]/div/section[1]/div[1]/div['+str(team_num)+']/div[2]').text
    
    return team_name

# 쿼터 점수가져오는 함수(team num : 1(홈팀), 2(어웨이팀), qt : 2~7(1~4쿼터, 연장, 총 점수))
def data_score(team_num, qt):
    score = driver.find_element(By.XPATH, '//*[@id="container"]/main/div[3]/div/section[1]/div[2]/table/tbody/tr['+str(team_num)+']/td['+str(qt)+']').text
    
    return score

# Stat 가져오는 함수(team num : 1(홈팀), 2(어웨이팀), stat_num : 1~14(2점슛 ~ 3점슛 성공률))
def data_stat(team_num, stat_num):
    if team_num==1:
        stat = driver.find_element(By.XPATH, '//*[@id="container"]/main/div[3]/div/section[2]/section[4]/table/tbody/tr['+str(stat_num)+']/td[1]/span[2]').text
    elif team_num==2:
        stat = driver.find_element(By.XPATH, '//*[@id="container"]/main/div[3]/div/section[2]/section[4]/table/tbody/tr['+str(stat_num)+']/td[3]/span[1]').text
    else:
        pass
    
    return stat
```

```python

//*[@id="container"]/main/div[3]/div/section[1]/div[2]/table/thead/tr/th[3]
```

```python
### 표를 만들기
### 크게 세 부분으로 추출
### 1. 경기 날짜, 팀 이름, 그 날의 경기 넘버, 홈/어웨이
### 2. 1~4 쿼터, 연장, 총 점수, Win/Lost
### 3. 경기 기록(2점, 3점 등)

def game_df(game_num):
    
    ## Part1. 경기날짜, 경기 번호, 팀이름, 홈/어웨이
    ## 홈팀의 인덱스는 1, 어웨이팀의 인덱스는 2로 지정하여 데이터 프레임 생성
    df = pd.DataFrame(columns=['Date', 'Game_num', 'Team', 'Home/Away'], index=[1,2]) # 컬럼과 인덱스 생성
    df['Date'] = data_date() # 경기 날짜 가져오는 함수 사용
    df['Game_num'] = game_num # game_num은 임의로 나중에 넣을 값(그 날의 경기 순서)
    
    for team_num in range(1,3): # 1=홈팀, 2=어웨이팀
        df['Team'][team_num] = data_team_name(team_num) # 행의 값=팀(홈/어웨이) 구분, 팀이름 가져오는 함수 사용
        if team_num == 1: # 홈/어웨이 넣어줄 것
            df['Home/Away'][team_num] = "Home"
        elif team_num == 2:
            df['Home/Away'][team_num] = "Away"
        else:
            pass
    
    ## Part2. 1Q~4Q, 연장, 총 점수 구하는 표
    score_name_list = [] # 점수 컬럼의 이름을 넣을 빈 리스트 생성
    score_home_list = [] # 홈팀의 점수 넣을 빈 리스트 생성
    score_away_list = [] # 어웨이팀의 점수 넣을 빈 리스트 생성
    
    for qt in range(2,8): # qt → 2~7 : 1~4/연장/총 점수
        score_name = driver.find_element(By.XPATH, '//*[@id="container"]/main/div[3]/div/section[1]/div[2]/table/thead/tr/th['+str(qt)+']').text
        score_name_list.append(score_name)
        
        for team_num in range(1,3):
            if team_num == 1:
                score_home = data_score(team_num, qt) # data_score 함수 활용
                score_home_list.append(score_home) # 빈리스트에 score_home 값 이어붙이기
            
            elif team_num ==2:
                score_away = data_score(team_num, qt) # data_score 함수 활용
                score_away_list.append(score_away) # 빈리스트에 score_away 값 이어붙이기
    
    score_df = pd.DataFrame(columns=score_name_list, index=[1,2], data=[score_home_list, score_away_list])
    # 스코어 데이터프레임 생성, 컬럼=각 점수 컬럼 이름, 인덱스=1,2, 데이터=score_home_list, score_away_list
    
    
    ## Part2-2 결합, 승리여부 넣어주기
    merged_df = pd.merge(df, score_df, left_index=True, right_index=True)
    merged_df['Win/Lose'] = 0 # 임의값 0으로 칼럼 생성
    
    if merged_df['TOTAL'][1] > merged_df['TOTAL'][2]: # 값 여부에 따라 win/lost 넣어주기
        merged_df['Win/Lose'][1] = "Won"
        merged_df['Win/Lose'][2] = "Lost"
    else:
        merged_df['Win/Lose'][1] = "Lost"
        merged_df['Win/Lose'][2] = "Won"
    
    
    ## Part3. 경기기록 표 구하기
    stat_columns_list = ["2PM", "3PM", "TR", "AST", "STL", "BLK", "FOUL", "FB", "TO", "PM", "GD", "FT%", "2P%", "3P%"]
    stat_home_list = []
    stat_away_list = []
    
    for stat_num in range(1, 15): # team_num → 1:홈팀 2:어웨이팀, stat_num → 1~14 : 2점슛 성공~3점슛 성공률
        for team_num in range(1,3):
            if team_num == 1:
                stat_home = data_stat(1, stat_num) # data_stat 함수 활용
                stat_home_list.append(stat_home) # stat_home 값 이어붙이기

            elif team_num == 2:
                stat_away = data_stat(2, stat_num) # data_stat 함수 활용
                stat_away_list.append(stat_away) # stat_away 값 이어붙이기
    
    stat_df = pd.DataFrame(columns=stat_columns_list, index=[1,2], data=[stat_home_list, stat_away_list])
    # stat_df 데이터프레임 생성, 칼럼, 인덱스=1,2, 인덱스 1의 값은 홈팀, 2의 값은 어웨이팀 데이터
    
    ## Part3-2 결합
    merged_df_2 = pd.merge(merged_df, stat_df, left_index=True, right_index=True) # 인덱스 기준으로 Part2-2, Part3 결합
    
    
    return merged_df_2
```

```python
### 하루동안 모든 경기 데이터 추출하는 함수
def day():
    
    click_match_box() # 경기 목록 클릭하기
    time.sleep(1) # 1초 휴식
    game_len = int(match_len()) # 그 날의 총 경기수 입력
    click_match_box()
    
    day_df = pd.DataFrame() # 하루의 모든 경기 데이터가 들어갈 빈 데이터 프레임 생성
    
    for game_num in range(1, (game_len+1)):
        click_match_box() # 경기 목록 클릭
        time.sleep(1)
        click_match(game_num) # 경기 목록 안의 경기 클릭
        time.sleep(1)
        one_game_df = game_df(game_num) # 데이터 추출함수 활용
        day_df = pd.concat([day_df, one_game_df]) # 데이터 프레임 이어 붙이기
    
    return day_df
```

```python
day()
```

```python
### 모든 날짜 반복하기
total_df = pd.DataFrame() # 빈 데이터프레임 생성
while True:
    day_df = day() # day_df 변수 = 하루의 모든 경기 데이터프레임
    time.sleep(1)
    total_df = pd.concat([total_df, day_df]) # 이어붙이기
    click_next_day() # 다음 날짜 클릭
    time.sleep(1)
```
