```python
# 연결 재무제표 추출 함수
def consolidated_fs(code):
    base_url = "https://navercomp.wisereport.co.kr/v2/company/c1010001.aspx?cmp_cd={}".format(code) # encparam, id 값을 찾기 위한 url
    main_text = requests.get(base_url).text # 전체 본문 텍스트
    
    pattern_enc = re.compile("encparam: '(.+)'", re.IGNORECASE) # encparam을 찾기 위한 패턴
    pattern_id  = re.compile("id: '(.+?)'", re.IGNORECASE) # id를 찾기 위한 패턴
    result_encparam = pattern_enc.search(main_text).groups()[0] # encparam: 대응값 변수 지정
    result_id = pattern_id.search(main_text).groups()[0] # id: 대응값 변수 지정
    
    # 파라미터 값 지정
    parameter = {
    'cmp_cd': code,
    'fin_typ': '4', # 4: 연결재무제표 / 3: 개별재무제표
    'freq_typ': 'Y', # Y: 연간보고서 / Q: 분기별보고서
    'encparam': result_encparam,
    'id': result_id}
    
    # 요청 헤더 값 지정
    head = {
    'Accept': 'text/html, */*; q=0.01',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive',
    'Host': 'navercomp.wisereport.co.kr',
    'Referer': 'https://navercomp.wisereport.co.kr/v2/company/c1010001',
    'X-Requested-With': 'XMLHttpRequest'}
    
    # url 지정
    fs_url = 'https://navercomp.wisereport.co.kr/v2/company/ajax/cF1001.aspx?'
    
    r = requests.get(fs_url, params=parameter, headers = head) # 특정 종목의 재무제표 데이터 관련 호출
    fs_df = pd.read_html(r.text)[1] # pd.read_html로 변환 값 중 1번째 값
    
    return fs_df
```

```python
# 특정 종목 연결 또는 개별 재무제표 추출 함수
def financial_statement(code, division, name):
    base_url = "https://navercomp.wisereport.co.kr/v2/company/c1010001.aspx?cmp_cd={}".format(code) # encparam, id 값을 찾기 위한 url
    main_text = requests.get(base_url).text # 전체 본문 텍스트
    
    pattern_enc = re.compile("encparam: '(.+)'", re.IGNORECASE) # encparam을 찾기 위한 패턴
    pattern_id  = re.compile("id: '(.+?)'", re.IGNORECASE) # id를 찾기 위한 패턴
    result_encparam = pattern_enc.search(main_text).groups()[0] # encparam: 대응값 변수 지정
    result_id = pattern_id.search(main_text).groups()[0] # id: 대응값 변수 지정
    
    # 파라미터 값 지정
    parameter = {
    'cmp_cd': code,
    'fin_typ': division, # 4: 연결재무제표 / 3: 개별재무제표
    'freq_typ': 'Y', # Y: 연간보고서 / Q: 분기별보고서
    'encparam': result_encparam,
    'id': result_id}
    
    # 요청 헤더 값 지정
    head = {
    'Accept': 'text/html, */*; q=0.01',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive',
    'Host': 'navercomp.wisereport.co.kr',
    'Referer': 'https://navercomp.wisereport.co.kr/v2/company/c1010001',
    'X-Requested-With': 'XMLHttpRequest'}
    
    # url 지정
    fs_url = 'https://navercomp.wisereport.co.kr/v2/company/ajax/cF1001.aspx?'
    
    # HTTP 조회
    r = requests.get(fs_url, params=parameter, headers = head) # 특정 종목의 재무제표 데이터 관련 호출
    fs_df = pd.read_html(r.text)[1] # pd.read_html로 변환 값 중 1번째 값
    
    # 추출한 데이터 프레임 정리
    fs_df_trans = fs_df.transpose()
    fs_df_trans = fs_df_trans.reset_index()
    fs_df_trans.columns = fs_df_trans.loc[0]
    fs_df_trans.drop([0], axis=0, inplace=True)
    
    # 컬럼 추가, 정보 입력
    fs_df_trans['종목코드'] = code
    fs_df_trans['종목명'] = name
    
    return fs_df_trans

# 연결+별도 재무제표 추출 함수
def total_fs(code, name):
    total_df = pd.concat([financial_statement(code, '4', name),
                         financial_statement(code, '3', name)], axis=0)
    
    return total_df
```

```python
# company에 등록된 모든 상장사 반복하여 재무제표 추출 합치기

# 이어 붙일 최종 데이터프레임 만들기
final_df = pd.DataFrame() 

# company 리스트 1번째 값부터 마지막까지
for count, index_num in enumerate(company[1:]):
    
    # 20의 배수때마다 진행상황 카운트 출력
    if count % 20 == 0:
        print(count)
        # 20의 배수마다 랜덤하게 2~3초 쉬기
        time.sleep(random.uniform(2, 4))
    
    # 종목코드, 종목명 변수로 지정
    company_code = index_num.split(",")[1].replace('"', '')
    company_name = index_num.split(",")[3].replace('"', '')
    
    # 출력한 데이터 프레임과 이전의 데이터 프레임 합치기
    try:
        final_df = pd.concat([final_df, total_fs(company_code, company_name)])
    except:
        pass
```
